```python
#Step 1. Import the necessary libraries
import pandas as pd
import random as rn
import numpy as np

#Step 2. Create 3 differents Series, each of length 100, as follows:
#length of series
length = 100
#Generating random series
#https://numpy.org/doc/stable/reference/random/generated/numpy.random.randint.html
#The first a random number from 1 to 4
series1 = pd.Series(np.random.randint(1,4,size=length))
#The second a random number from 1 to 3
series2 = pd.Series(np.random.randint(1,3,size=length))
#The third a random number from 10,000 to 30,000
series3 = pd.Series(np.random.randint(10000,30000,size=length))
#Step 3. Create a DataFrame by joinning the Series by column
dataFrameOfSeries =  pd.DataFrame({'series1':series1, 'series2': series2, 'series3': series3})
dataFrameOfSeries
```

```python
#Step 4. Change the name of the columns to bedrs, bathrs, price_sqr_meter
dataFrameOfSeries.columns = ['bedrs', 'bathrs', 'price_sqr_meter']
dataFrameOfSeries
```

```python
#Step 5. Create a one column DataFrame with the values of the 3 Series and assign it to 'bigcolumn'
#Reference: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.concat.html
bigcolumn = pd.concat([series1,series2,series3])
print(bigcolumn)
#Step 6. Ops it seems it is going only until index 99. Is it true?
print('\nStep 6: Yes, it is going until 99')
#Step 7. Reindex the DataFrame so it goes from 0 to 299
print('\nStep 7: Reindexed from 0 to 299')
bigcolumn = bigcolumn.reset_index(drop=True)
print(bigcolumn)
```

```python
#Step 1. Import the necessary libraries
import pandas as pd
import datetime
#Step 2. Import the dataset from this address
#Step 3. Assign it to a variable called data and replace the first 3 columns by a proper datetime index.
#Reference https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.read_csv.html
url = r'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/Wind_Stats/wind.data'
windData = pd.read_csv(url , sep  ="\s+",parse_dates={ 'Date': ['Yr', 'Mo', 'Dy'] } )
print('Step 3. Wind data read with the first 3 columns combined')
windData
```

```python
#Step 4. Year 2061? Do we really have data from this year? Create a function to fix it and apply it.
#Reference: class notes for lambda function and https://www.programiz.com/python-programming/datetime for datetime.date
windData['Date'] = windData['Date'].apply(lambda x: datetime.date(x.year-100,x.month,x.day) if x.year >=2000 else x)
print('\nStep 4. Wind data with dates fixed')
windData
```

```python
#Step 5. Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns].
windData['Date'] = pd.to_datetime(windData['Date'])
windData = windData.set_index('Date')
print('\nStep 5. Wind data with date column index and in datetime64[ns] format')
windData
```

```python
#Step 6. Compute how many values are missing for each location over the entire record.They should be ignored in all calculations below.
missingValues = windData.isnull().sum()
print('\nStep 6. Count of all the missing values for each location in Wind data')
missingValues
```

```python
#Step 7. Compute how many non-missing values there are in total.
nonMissingValues = windData.count()
print('\nStep 7. Count of all the non missing values for each location in Wind data')
nonMissingValues
```

```python
#Step 8. Calculate the mean windspeeds of the windspeeds over all the locations and all the times.
#A single number for the entire dataset.
m = windData.mean(axis = 0) #Mean of each column
sumA = m.sum() #sum of means of all the coumns
countA = m.count() #count of the columns
mean = sumA / countA #calculating mean
print('\nStep 8. Mean windspeeds of the windspeeds over all the locations and all the times')
mean
```

```python
#Step 9. Create a DataFrame called loc_stats and calculate the min, max and mean windspeeds and standard deviations of the windspeeds at each location over all the days
#Reference class notes
#loc_stats = windData.describe() #This shows all the summary statistics and we require only min, max, mean and standard deviation
loc_stats = pd.DataFrame({'MIN':  windData.min(), 'MAX': windData.max(), 'MEAN': windData.mean(), 'STD': windData.std() })
print('\nStep 9. Min, Max, Mean and Standard Deviation of the data')
loc_stats
```

```python
#Step 10. Create a DataFrame called day_stats and calculate the min, max and mean windspeed and standard deviations of the windspeeds across all the locations at each day.
day_stats = pd.DataFrame({'MIN': windData.min(axis = 1), 'MAX': windData.max(axis = 1), 'MEAN': windData.mean(axis = 1), 'STD': windData.std(axis = 1) })
print('\nStep 10. Min, Max, Mean and Standard Deviation of the data for each day')
day_stats
```

```python
#Step 11. Find the average windspeed in January for each location.
avgWind = windData.loc[windData.index.month == 1].mean()
print('\nStep 11. Average windspeed in January for each location')
avgWind
```

```python
#Step 12. Downsample the record to a yearly frequency for each location.
#reference: https://pandas.pydata.org/pandas-docs/version/0.23.1/timeseries.html#timeseries-offset-aliases
windDataIndexedAnually = windData.groupby(windData.index.to_period('A')).mean()
windDataIndexedAnually
```

```python
#Step 13. Downsample the record to a monthly frequency for each location.
windDataIndexedMonthly = windData.groupby(windData.index.to_period('M')).mean()
windDataIndexedMonthly
```

```python
#Step 14. Downsample the record to a weekly frequency for each location.
windDataIndexedWeekly = windData.groupby(windData.index.to_period('W')).mean()
windDataIndexedWeekly
```
